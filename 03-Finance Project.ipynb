## Reading the data

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2024,1,1)
```

```python
#Bank of America
BAC = data.DataReader('BAC', data_source='stooq', start=start, end=end)

#CitiGroup
C = data.DataReader('C', data_source='stooq', start=start, end=end)

#Goldman Sachs
GS = data.DataReader('GS', data_source='stooq', start=start, end=end)

#JPMorgan Chase
JPM = data.DataReader('JPM', data_source='stooq', start=start, end=end)

#Morgan Stanley
MS = data.DataReader('MS', data_source='stooq', start=start, end=end)

#Wells Fargo
WFC = data.DataReader('WFC', data_source='stooq', start=start, end=end)
```

## Organizing the data

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

** Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list. Also pay attention to what axis you concatenate on.**

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], keys=tickers, axis=1)
```

```python
bank_stocks.head()
```

```python
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
```

```python
bank_stocks.head()
```

## Explore the returns of each bank's stock. 
Returns are typically defined by:

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

```python
for ticker in tickers:
    name = ticker + " Returns"
    returns[name] = bank_stocks[ticker]['Close'].pct_change() #get the returns for the bank
```

```python
returns
```

### Visualize the returns

### Histogram of Morgan Stanley in 2023 Vs. 2008
